```python
import pandas as pd
import numpy as np
from tqdm import tqdm

# Definir el tamaÃ±o del lote
batch_size = 10000

# Cargar los datos en lotes
patient_data = pd.read_csv('PatientCorePopulatedTable.txt', sep='\t', chunksize=batch_size)
labs_data = pd.read_csv('LabsCorePopulatedTable.txt', sep='\t', chunksize=batch_size)
diagnosis_data = pd.read_csv('AdmissionsDiagnosesCorePopulatedTable.txt', sep='\t', chunksize=batch_size)

# Preprocesar y guardar los datos por lotes
batch_num = 1
for patient_batch in tqdm(patient_data, desc='Processing Patient Data'):
    patient_batch['PatientDateOfBirth'] = pd.to_datetime(patient_batch['PatientDateOfBirth'])
    current_date = pd.to_datetime('today')
    patient_batch['Age'] = current_date.year - patient_batch['PatientDateOfBirth'].dt.year
    
    for labs_batch in tqdm(labs_data, desc='Processing Labs Data'):
        merged_batch = pd.merge(labs_batch, patient_batch, on='PatientID', how='inner')
        
        for diag_batch in tqdm(diagnosis_data, desc='Processing Diagnosis Data'):
            merged_final_batch = pd.merge(merged_batch, diag_batch[['PatientID', 'PrimaryDiagnosisCode', 'PrimaryDiagnosisDescription']], on='PatientID', how='inner')
            
            # Filtrar las columnas relevantes
            features = ['AdmissionID', 'PatientID', 'LabName', 'LabValue', 'PatientGender', 'Age', 'PatientRace']
            labels = ['PrimaryDiagnosisCode', 'PrimaryDiagnosisDescription']
            merged_final_batch = merged_final_batch[features + labels]
            
            # Guardar el lote procesado en un archivo CSV
            merged_final_batch.to_csv(f'processed_batch_{batch_num}.csv', index=False)
            batch_num += 1
```

```python
df = merged_final_batch.copy()

# Crear un DataFrame pivotado
df_pivot = df.pivot_table(
    index=['AdmissionID', 'PatientGender', 'Age', 'PatientRace'],
    columns='LabName',
    values='LabValue',
    aggfunc='first'
).reset_index()

# Unir diagnÃ³sticos por AdmissionID
diagnosis_info = df[['AdmissionID', 'PrimaryDiagnosisCode', 'PrimaryDiagnosisDescription']].drop_duplicates()
diagnosis_info_grouped = diagnosis_info.groupby('AdmissionID').agg(list).reset_index()

# Unir diagnÃ³sticos con el DataFrame pivotado
df_final = pd.merge(df_pivot, diagnosis_info_grouped, on='AdmissionID')
```

```python
import pandas as pd

df = merged_final_batch.copy()

# Crear un DataFrame pivotado
df_pivot = df.pivot_table(
    index=['AdmissionID', 'PatientID', 'PatientGender', 'Age', 'PatientRace'],
    columns='LabName',
    values='LabValue',
    aggfunc='first'
).reset_index()

# Unir diagnÃ³sticos por AdmissionID
diagnosis_info = df[['AdmissionID', 'PatientID', 'PrimaryDiagnosisCode', 'PrimaryDiagnosisDescription']].drop_duplicates()

# Agrupar diagnÃ³sticos por PatientID
diagnosis_grouped = diagnosis_info.groupby('PatientID').agg({
    'PrimaryDiagnosisCode': lambda x: ', '.join(set(x)),
    'PrimaryDiagnosisDescription': lambda x: ', '.join(set(x))
}).reset_index()

# Unir diagnÃ³sticos con el DataFrame pivotado usando PatientID
df_final = pd.merge(df_pivot, diagnosis_grouped, on='PatientID')

# Guardar el DataFrame final en un archivo CSV
df_final.to_csv('processed_data.csv', index=False)

# Mostrar el DataFrame final
print(df_final)
```

```python
diagnosis_data_df = pd.read_csv('AdmissionsDiagnosesCorePopulatedTable.txt', sep='\t')

# Unir diagnÃ³sticos con df_final usando PatientID y AdmissionID
df_final_with_diagnosis = pd.merge(df_final, diagnosis_data_df, on=['PatientID', 'AdmissionID'], how='left')

# Mostrar el DataFrame final con diagnÃ³sticos aÃ±adidos
print(df_final_with_diagnosis.head())

# Guardar el DataFrame final con diagnÃ³sticos en un archivo CSV
df_final_with_diagnosis.to_csv('processed_data_with_diagnosis.csv', index=False)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el DataFrame df_final_with_diagnosis (asegÃºrate de haberlo definido previamente)

# Seleccionar las columnas relevantes para el anÃ¡lisis de comorbilidades
diagnosis_columns = ['PrimaryDiagnosisCode_x', 'PrimaryDiagnosisDescription_x']

# Filtrar solo las filas donde ambas columnas no son NaN
diagnosis_pairs = df_final_with_diagnosis[diagnosis_columns].dropna()

# Conteo de las comorbilidades mÃ¡s frecuentes
comorbidity_counts = diagnosis_pairs.groupby(diagnosis_columns).size().reset_index(name='count')

# Ordenar por frecuencia de comorbilidad (opcional)
comorbidity_counts = comorbidity_counts.sort_values(by='count', ascending=False)

# Mostrar las combinaciones mÃ¡s frecuentes de diagnÃ³sticos
print("Combinaciones mÃ¡s frecuentes de diagnÃ³sticos:")
print(comorbidity_counts.head(10))
print()

# VisualizaciÃ³n: Diagrama de calor (heatmap) de comorbilidades
# Crear una matriz de comorbilidad (0 o 1 dependiendo de la existencia de la combinaciÃ³n)
comorbidity_matrix = pd.crosstab(diagnosis_pairs['PrimaryDiagnosisCode_x'], diagnosis_pairs['PrimaryDiagnosisDescription_x'])

plt.figure(figsize=(12, 10))
sns.heatmap(comorbidity_matrix, cmap="YlGnBu", annot=True, fmt="d")
plt.title('Matriz de Comorbilidad entre DiagnÃ³sticos')
plt.xlabel('PrimaryDiagnosisCode_y')
plt.ylabel('PrimaryDiagnosisCode_x')
plt.show()
```
